# Import Essential Libraries

```
# python visualization libraries
import pandas as pd
import numpy as np

import math
import json


import plotly.express as px
import plotly.graph_objects as go

# for hierarchical data
import networkx as nx
```

# Load Data from GitHub

```
# ref: https://stackoverflow.com/questions/32400867/pandas-read-csv-from-url

url = 'https://raw.githubusercontent.com/lindyhernandez/IST526/main/HW5/IST_Faculty'

# pandas call to read csv file 
df = pd.read_csv(url)

# quickly show the dataframe
display(df.head())
display(df.tail())
```

# Basic Sunburst Tree

```
fig = px.sunburst(
  df,   
  parents = 'Parent',
  names = 'Item and Group',
  values='Weight'
)

fig.update_layout(
  title_text="Sunburst Diagram", 
  font_size=12
)

fig.show()
```

# Treemaps

```
fig = px.treemap(
  df,   
  parents = 'Parent',
  names = 'Item and Group',
  values='Weight'
)

fig.update_layout(
  title_text="Treemap Diagram", 
  font_size=12,
  margin = dict(t=50, l=25, r=25, b=25)
)

fig.update_traces(
  root_color="lightgrey"
)

fig.show()
```
